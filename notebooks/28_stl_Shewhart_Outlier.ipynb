```python
from io import StringIO
df = pd.read_csv(StringIO(raw_data))
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)

# ----------------------------
# 2. STL 분해
# ----------------------------
result = seasonal_decompose(df['Total'], model='additive', period=12)
resid = result.resid.dropna()
x = resid.index
y = resid.values

# ----------------------------
# 3. 슈하르트 관리도 기준선 계산
# ----------------------------
mean = np.mean(y)
std = np.std(y)
ucl = mean + 3 * std
lcl = mean - 3 * std

# ----------------------------
# 4. 잔차 + 관리도 시각화
# ----------------------------
plt.figure(figsize=(18, 5))
plt.plot(x, y, label='Residuals', linestyle='--', color='gray')
plt.axhline(mean, color='green', linestyle='-', label='Center Line (Mean)')
plt.axhline(ucl, color='red', linestyle='--', label='UCL (+3σ)')
plt.axhline(lcl, color='blue', linestyle='--', label='LCL (-3σ)')

# 이상치 표시
for i in range(len(y)):
    if y[i] > ucl or y[i] < lcl:
        plt.annotate(
            'Anomaly',
            (mdates.date2num(x[i]), y[i]),
            xytext=(30, 20),
            textcoords='offset points',
            color='red',
            arrowprops=dict(facecolor='red', arrowstyle='fancy')
        )

plt.title("Shewhart Control Chart (Residuals after STL)")
plt.xlabel("Date")
plt.ylabel("Residual")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# ----------------------------
# 5. 원 시계열에서 이상치 주석 추가
# ----------------------------
plt.rc('figure', figsize=(18,5))
plt.rc('font', size=15)
ax = df.plot()

for i in range(len(y)):
    date = x[i]
    if y[i] < lcl or y[i] > ucl:
        ax.annotate(
            'Anomaly',
            (date, df.loc[date, 'Total']),
            xytext=(30, 20),
            textcoords='offset points',
            color='red',
            arrowprops=dict(facecolor='red', arrowstyle='fancy')
        )

plt.title("Original Time Series with Detected Anomalies")
plt.xlabel("Date")
plt.ylabel("Total")
plt.grid(True)
plt.tight_layout()
plt.show()
```
