# # ✅ ARIMA(p, d, q) 모델을 활용한 테슬라 주가 예측

```python
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.arima.model import ARIMA
from pmdarima import auto_arima
from sklearn.metrics import mean_squared_error, r2_score

# 📌 1️⃣ 테슬라(TSLA) 주가 데이터 다운로드 (최근 2년치)
df = yf.download("TSLA", start="2022-01-01", end="2024-01-01")
df = df.reset_index()

# 📌 2️⃣ 학습 및 테스트 데이터 분할 (80% 학습, 20% 테스트)
train_size = int(len(df) * 0.8)
train, test = df.iloc[:train_size].copy(), df.iloc[train_size:].copy()

# 데이터 확인 (디버깅용)
print(f"Train DataFrame: {train.shape}")
print(f"Test DataFrame: {test.shape}")
print(train.head())

# 📌 3️⃣ 최적의 (p, d, q) 값 찾기
auto_model = auto_arima(train["Close"], seasonal=False, stepwise=True, trace=True)
best_p, best_d, best_q = auto_model.order
print(f"✅ 최적의 (p, d, q) 값: (p={best_p}, d={best_d}, q={best_q})")

# 📌 4️⃣ ARIMA 모델 학습
model = ARIMA(train["Close"], order=(best_p, best_d, best_q))
model_fit = model.fit()

# 📌 5️⃣ 예측 수행
train_pred = model_fit.predict(start=best_p, end=len(train) - 1, typ="levels")
test_pred = model_fit.forecast(steps=len(test))  # 롤링 예측이 아니라, 전체 예측

# 📌 6️⃣ 성능 평가
train_actual = train["Close"][best_p:]
test_actual = test["Close"]

train_mse = mean_squared_error(train_actual, train_pred)
test_mse = mean_squared_error(test_actual, test_pred)
train_r2 = r2_score(train_actual, train_pred)
test_r2 = r2_score(test_actual, test_pred)

print(f"📌 Training MSE: {train_mse:.4f}, R²: {train_r2:.4f}")
print(f"📌 Test MSE: {test_mse:.4f}, R²: {test_r2:.4f}")

# 📌 7️⃣ 결과 시각화 (전체 데이터 예측)
plt.figure(figsize=(12, 5))
plt.plot(df["Date"], df["Close"], label="Actual", color="black")
plt.plot(test["Date"], test_pred, label="Predicted (ARIMA Model)", linestyle="--", color="red")
plt.title("Tesla Stock Price Prediction using ARIMA Model")
plt.xlabel("Date")
plt.ylabel("Stock Price (USD)")
plt.xticks(rotation=45)
plt.legend()
plt.show()

# 📌 8️⃣ 결과 시각화 (훈련 데이터 예측 vs 실제값)
plt.figure(figsize=(12, 5))
plt.plot(train["Date"][best_p:], train_actual, label="Actual (Training)", color="black")
plt.plot(train["Date"][best_p:], train_pred, label="Predicted (Training)", color="blue", linestyle="--")
plt.title("Tesla Stock Price Prediction using ARIMA - Training Phase")
plt.xlabel("Date")
plt.ylabel("Stock Price (USD)")
plt.xticks(rotation=45)
plt.legend()
plt.show()

# 📌 9️⃣ 결과 시각화 (테스트 데이터 예측 vs 실제값)
plt.figure(figsize=(12, 5))
plt.plot(test["Date"], test_actual, label="Actual (Test)", color="black")
plt.plot(test["Date"], test_pred, label="Predicted (Test)", linestyle="--", color="red")
plt.title("Tesla Stock Price Prediction using ARIMA - Test Phase")
plt.xlabel("Date")
plt.ylabel("Stock Price (USD)")
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

# ✅ 최적 (p, d, q) 탐색 + ARIMA 모델 학습 + 롤링 예측 코드

✅ 개선된 기능
최적의 (p, d, q) 값 자동 탐색
pmdarima.auto_arima()를 사용하여 최적의 파라미터를 자동으로 찾음.
ARIMA 모델 학습 및 전체 예측
model_fit.predict()로 훈련 및 테스트 데이터 예측 수행.
롤링 예측 (One-step rolling forecast)
테스트 데이터를 한 스텝씩 예측하는 방식 적용.
성능 평가 및 시각화
MSE(Mean Squared Error) 및 R²(설명력) 계산
훈련 데이터, 테스트 데이터에 대한 예측 결과 시각화.

```python
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.arima.model import ARIMA
from pmdarima import auto_arima
from sklearn.metrics import mean_squared_error, r2_score

# 📌 1️⃣ 테슬라(TSLA) 주가 데이터 다운로드 (최근 2년치)
df = yf.download("TSLA", start="2022-01-01", end="2024-01-01")

# ✅ 다중 인덱스 제거 (필수)
df.columns = df.columns.droplevel(1)  # MultiIndex 제거
df = df.reset_index()  # 인덱스 초기화

# 📌 ✅ 데이터프레임 컬럼 확인 (디버깅)
print("🔍 DataFrame Columns:", df.columns)
print(df.head())

# 📌 2️⃣ 학습 및 테스트 데이터 분할 (80% 학습, 20% 테스트)
train_size = int(len(df) * 0.8)
train, test = df.iloc[:train_size].copy(), df.iloc[train_size:].copy()

# 📌 3️⃣ 최적의 (p, d, q) 값 찾기
auto_model = auto_arima(train["Close"], seasonal=False, stepwise=True, trace=True)
best_p, best_d, best_q = auto_model.order
print(f"✅ 최적의 (p, d, q) 값: (p={best_p}, d={best_d}, q={best_q})")

# 📌 4️⃣ ARIMA 모델 학습 (전체 예측용)
model = ARIMA(train["Close"].astype(float), order=(best_p, best_d, best_q))  # ✅ float 변환
model_fit = model.fit()

# 📌 5️⃣ 예측 수행
train_pred = model_fit.predict(start=best_p, end=len(train) - 1, typ="levels")
test_pred = model_fit.forecast(steps=len(test))

# 📌 6️⃣ 성능 평가
train_actual = train["Close"].iloc[best_p:]  # 실제값
test_actual = test["Close"]

train_mse = mean_squared_error(train_actual, train_pred)
test_mse = mean_squared_error(test_actual, test_pred)
train_r2 = r2_score(train_actual, train_pred)
test_r2 = r2_score(test_actual, test_pred)

print(f"📌 Training MSE: {train_mse:.4f}, R²: {train_r2:.4f}")
print(f"📌 Test MSE: {test_mse:.4f}, R²: {test_r2:.4f}")

# 📌 7️⃣ 롤링 예측 (One-step rolling forecast)
rolling_predictions = []
history = list(train["Close"].dropna().astype(float))  # ✅ float 변환 및 NaN 제거

# 📌 ✅ history 리스트 확인 (디버깅)
print("🔍 Initial History Values:", history[:10])

for t in range(len(test)):
    history_array = np.array(history, dtype=float)  # ✅ NumPy 배열 변환
    model = ARIMA(history_array, order=(best_p, best_d, best_q))
    model_fit = model.fit()
    pred = model_fit.forecast(steps=1)[0]  # 한 스텝 예측
    rolling_predictions.append(pred)

    # ✅ `float()` 변환 후 추가 (디버깅)
    try:
        history.append(float(test["Close"].iloc[t]))  # float 변환하여 추가
    except ValueError:
        print(f"❌ 변환 실패: test['Close'].iloc[{t}] = {test['Close'].iloc[t]}")  # 에러 메시지 출력

# 📌 8️⃣ 결과 시각화 (전체 데이터 예측)
plt.figure(figsize=(12, 5))
plt.plot(df["Date"], df["Close"], label="Actual", color="black")
plt.plot(test["Date"], rolling_predictions, label="Predicted (ARIMA Model)", linestyle="--", color="red")
plt.title("Tesla Stock Price Prediction using ARIMA Model (Full Forecast)")
plt.xlabel("Date")
plt.ylabel("Stock Price (USD)")
plt.xticks(rotation=45)
plt.legend()
plt.show()

# 📌 9️⃣ 결과 시각화 (훈련 데이터 예측 vs 실제값)
plt.figure(figsize=(12, 5))
plt.plot(train["Date"].iloc[best_p:], train_actual, label="Actual (Training)", color="black")
plt.plot(train["Date"].iloc[best_p:], train_pred, label="Predicted (Training)", color="blue", linestyle="--")
plt.title("Tesla Stock Price Prediction using ARIMA - Training Phase")
plt.xlabel("Date")
plt.ylabel("Stock Price (USD)")
plt.xticks(rotation=45)
plt.legend()
plt.show()

# 📌 🔟 결과 시각화 (테스트 데이터 예측 vs 실제값 - 롤링 예측)
plt.figure(figsize=(12, 5))
plt.plot(test["Date"], test_actual, label="Actual (Test)", color="black")
plt.plot(test["Date"], rolling_predictions, label="Rolling Predicted (Test)", linestyle="--", color="green")
plt.title("Tesla Stock Price Prediction using ARIMA - Rolling Forecast")
plt.xlabel("Date")
plt.ylabel("Stock Price (USD)")
plt.xticks(rotation=45)
plt.legend()
plt.show()
```
